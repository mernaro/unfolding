```python
import pandas as pd

def concat_last_rows(csv_files, name_index):
    last_rows = []
    for file in csv_files:
        df = pd.read_csv(file)
        last_rows.append(df.tail(1))
    result = pd.concat(last_rows, ignore_index=True)
    result.drop('Image_ID', axis=1, inplace=True)
    result.index = name_index
    return result

def simplify_print_data(df):
    psnr_diff = df['PSNR_Output'] - df['PSNR_Input']
    mse_diff =  df['MSE_Output'] - df['MSE_Input']
    mae_diff = df['MAE_Output'] - df['MAE_Input']
    maxae_diff = df['MaxAE_Output'] - df['MaxAE_Input']
    result = pd.DataFrame({
        'PSNR_Diff':psnr_diff,
        'MSE_Diff':mse_diff,
        'MAE_Diff':mae_diff,
        'MaxAE_Diff':maxae_diff
    }, index=df.index)
    return result

experiments_dir = "/projects/memaro/rpujol/unfolding/experiments/"
name_file = 'metrics.csv'
```

```python
date = "2026-02-0"
expe_dir_act = experiments_dir + date
csv_list = [expe_dir_act+"5_15-49/"+name_file,expe_dir_act+"5_16-57/"+name_file,expe_dir_act+"6_11-43/"+name_file]
name_index = ["0.0001","0.01","0"]
df_last = concat_last_rows(csv_list, name_index)
display(df_last)
df_simplify = simplify_print_data(df_last)
display(df_simplify)
stats = df_simplify.agg(['mean', 'std'])
display(stats)
```
